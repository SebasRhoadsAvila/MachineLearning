```python
import requests
import json

def buscar_chiste_chuck_norris(palabra):
    try:
        # Realizar la solicitud a la API con la palabra ingresada
        url = f"https://api.chucknorris.io/jokes/search?query={palabra}"
        respuesta = requests.request("GET", url)
        # Cargar el contenido de la respuesta y extraer los chistes
        contenido = json.loads(respuesta.text)
        chistes = contenido['result']
        
        # Verificar si se encontraron chistes
        if chistes:
            # Devolver el primer chiste encontrado
            return chistes[0]['value']
        else:
            # Mensaje si no se encuentran chistes con la palabra ingresada
            return "No se encontró un chiste con la palabra introducida."
    except Exception as e:
        # Manejo de errores en la solicitud
        return f"Ocurrió un error al buscar el chiste: {str(e)}"

# Ejemplo de uso de la función
buscar_chiste_chuck_norris("computer")
```
