# Problema 1 

### Escribir una función para generar N muestras de una distribución normal, y generar un histograma

```python
def dist_normal(n):
    muestra=np.random.randn(n)
    plt.hist(muestra, bins = 100)
    plt.show()
```

```python
dist_normal(100000)
```

# Problema 2: B

### Dados dos vectores, escribir el código que imprime los números que se repiten en ambos

```python
def valores_repetidos(v1,v2):
    resultado = []
    for i in v1:
        if i in v2 and i not in resultado:
            resultado.append(i)
    return resultado
```

```python
v1= [1,3,5,8,6,4]
v2= [1,10,9,4]
valores_repetidos(v1,v2)
```
